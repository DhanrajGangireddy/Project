# Feature Extraction

```python
def makeTokens(f):
    tkns_BySlash = str(f.encode('utf-8')).split('/') 
    total_Tokens = []
    for i in tkns_BySlash:
        tokens = str(i).split('-')  
        for token in tokens:
            temp_Tokens = str(token).split('.') + str(token).split('?') + str(token).split('//')
            total_Tokens.extend(temp_Tokens)
    total_Tokens = list(set(total_Tokens))  
    if 'com' in total_Tokens:
        total_Tokens.remove('com')
    if 'www' in total_Tokens:
        total_Tokens.remove('www')
    return total_Tokens
```

# Divide the Data Train and Test 80:20 ratio

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split

X = df['url']
y = df['status']


tfidf_vectorizer = TfidfVectorizer(tokenizer=makeTokens)
X_url_num = tfidf_vectorizer.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X_url_num, y, test_size=0.2, random_state=0)
```

# Scaling

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler(with_mean=False)

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

scaling_values = scaler.scale_[:11]
scaling_values
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
logit = LogisticRegression()

logit.fit(X_train_scaled, y_train)

y_pred = logit.predict(X_test_scaled)

logistic_accuracy = accuracy_score(y_test, y_pred)
print("Accuracy: {:.2f}".format(logistic_accuracy))
```

```python
from sklearn.metrics import classification_report

report = classification_report(y_test, y_pred)
print("Classification Report:\n", report)
```

```python
from sklearn.metrics import confusion_matrix

conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", conf_matrix)
```

```python
plt.figure(figsize=(6, 6))
sns.heatmap(conf_matrix, annot=True, cmap='Greens', fmt='g', cbar=False)
plt.title('Confusion Matrix')
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.show()
```

# Algorithms

```python
from sklearn.tree import DecisionTreeClassifier

decision = DecisionTreeClassifier()

decision.fit(X_train_scaled,y_train)

y_pred = decision.predict(X_test_scaled)

Decision_accuracy = accuracy_score(y_test,y_pred)
print("Accuracy: {:.2f}".format(Decision_accuracy))
```

```python
from sklearn.ensemble import RandomForestClassifier

random = RandomForestClassifier()

random.fit(X_train_scaled,y_train)

y_pred = random.predict(X_test_scaled)

Random_accuracy = accuracy_score(y_test,y_pred)
print("Accuracy: {:.2f}".format(Random_accuracy))
```

```python
from sklearn.svm import SVC

sup = SVC()

sup.fit(X_train_scaled,y_train)

y_pred = sup.predict(X_test_scaled)

SVC_accuracy = accuracy_score(y_test,y_pred)
print("Accuracy: {:.2f}".format(SVC_accuracy))
```

```python
from sklearn.ensemble import GradientBoostingClassifier

gb = GradientBoostingClassifier()

gb.fit(X_train_scaled,y_train)

y_pred = gb.predict(X_test_scaled)

Gradient_accuracy = accuracy_score(y_test,y_pred)
print("Accuracy: {:.2f}".format(Gradient_accuracy))
```

```python
models = ['Logistic Regression', 'Decision Tree', 'Random Forest', 'SVM', 'GradientBoosting']
accuracies = [logistic_accuracy, Decision_accuracy, Random_accuracy, SVC_accuracy, Gradient_accuracy]

plt.figure(figsize=(8,6))
plt.bar(models, accuracies, color=['blue', 'green', 'red', 'orange', 'purple'])
plt.title('Accuracy Comparison of Five Models')
plt.xlabel('Model')
plt.ylabel('Accuracy')
plt.ylim(0, 1)  
plt.grid(axis='y')  
plt.yticks([i/10 for i in range(11)])

plt.show()
```
